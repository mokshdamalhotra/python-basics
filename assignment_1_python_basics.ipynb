```python
#Q1: L=[11,12,13,14]
L=[11,12,13,14]
#(i) WAP to add 50 and 60 to L.
L.append(50)
L.append(60)
#(ii) WAP to remove 11 and 13from L.
L.remove(11)
L.remove(13)
#(iii) WAP to sort L in ascending order.
L.sort()
print("list in ascending order:",L)
#(iv) WAP to sort L in descending order.
L.sort(reverse=True)
print("List in descending order:",L)
#(v) WAP to search for 13 in L.
try:
  num=L.index(13)
  print("13 is present at index: ", num)
except:
  print("13 is not in list")
#(vi) WAP to count the number of elements present in L.
print("number of list elements:",len(L))
#(vii) WAP to sum all the elements in L.
sum=0
for i in L:
  sum+=i
print("sum of list elements:",sum)
#(viii) WAP to sum all ODD numbers in L.
#(ix) WAP to sum all EVEN numbers in L
sum_odd=0
sum_even=0
for m in L:
  if m%2==0:
    sum_even+=m
  else:
    sum_odd+=m
print("sum of even numbers:", sum_even, "\nsum of odd numbers:", sum_odd)
#(x) WAP to sum all PRIME numbers in L.
sum_prime=0
for n in L:
  for x in range(2,(n//2)+1):
    if n%x==0:
      break
    else:
      print("prime number:", n)
      sum_prime+=n
      continue
print("sum of prime numbers:",sum_prime)
#(xi) WAP to clear all the elements in L.
L.clear()
print("list now:", L)
#(xii) WAP to delete L
del(L)
try:
  print("the list now:", L)
except:
  print("list no longer exists")
```

```python
#Q5. WAP to create two lists of 10 random numbers between 10 and 30; Find
import random as r
L1,L2=[],[]
for i in range (1,11):
  L1.append(r.randrange(10,30))
  L2.append(r.randrange(10,30))
  i+=1
print("list 1 :", L1, "\nlist 2:", L2)
common=[]
count=0
for m in L1:
  for n in L2:
    if m!=n:
      continue
    else:
      if m in common:
        count+=1
        continue
      else:
        count+=2
        common.append(m)
print("-----------------------------------------------------------------------")
#(i) Common numbers in the two lists

print("number of common numbers:", len(common))
print("-----------------------------------------------------------------------")
#(ii) Unique numbers in both the list

print("number of unique numbers:", len(L1)+len(L2)-count)
print("-----------------------------------------------------------------------")
#(iii) Minimum in both the list

print("Minimum is list 1:", min(L1), "\nminimum in list 2:", min(L2))
print("-----------------------------------------------------------------------")
#(iv) Maximum in both the list

print("maximum is list 1:", max(L1), "\nmaximum in list 2:", max(L2))
print("-----------------------------------------------------------------------")
#(v) Sum of both the lists
sum=0
for j in L1:
  sum+=j
for k in L2:
  sum+=k
print("Sum of both the lists:", sum)
print("-----------------------------------------------------------------------")
```

```python
#Q6. WAP to create a list of 100 random numbers between 100 and 900.
import random as r
list1=[]
for i in range(1,101):
  list1.append(r.randrange(100,900))
print("the list is:",list1)
#Count and print the:
#(i) All odd numbers
list_even,list_odd=[],[]
print("-----------------------------------------------------------------------")
for m in list1:
  if m%2==0:
    list_even.append(m)
  else:
    list_odd.append(m)
print("odd numbers:", list_odd,"\nnumber of odd numbers:", len(list_odd))
#(ii) All even numbers
print("-----------------------------------------------------------------------")
print("even numbers:", list_even,"\nnumber of even numbers:", len(list_even))
#(iii) All prime numbers
print("-----------------------------------------------------------------------")
ct=[]
for x in list1:
  factor=0
  for y in range(2,(x//2)+1):
    if x%y==0:
      factor+=1
      break
    else:
      continue
  if factor==0:
    ct.append(x)
print("prime numbers:", ct,"\nnumber of prime numbers:", len(ct))
print("-----------------------------------------------------------------------")
```

```python
#Q7. D is a dictionary defined as D={1:"One",2:"Two",3:"Three",4:"Four", 5:"Five"}.
#WAP to read all the keys and values from dictionary and write to the file in the given below format.
#Key1, Value1
#Key2, Value2
#Key3, Value3
D={1:"One",2:"Two",3:"Three",4:"Four", 5:"Five"}
print("dictionary keys:", D.keys(),"\ndictionary values:",D.values())
fpw=open('dictionary_file.txt','w')
for i in D.keys():
  str_i=str(i)
  key_i=str(D[i])
  fpw.write(str_i+','+key_i+"\n")
fpw.close()
```

```python
#Q10. Write to the file all prime numbers between 600 and 800
fp=open('prime_numbers.txt','w')
fp.write("list of prime numbers between 600 and 800\n")
for i in range(600,801):
  factor=0
  for j in range(2,(i//2)+1):
    if i%j==0:
      factor+=1
      break
    else:
      continue
  if factor==0:
    str_num=str(i)
    fp.write(str_num+"\n")
fp.close()
```
